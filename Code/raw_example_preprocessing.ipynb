```python
import pandas as pd

# Open the spreadsheet by URL
spreadsheet_url = 'https://docs.google.com/spreadsheets/d/1g7wgv7mXCxOPtdXBBvgATwKyysThCbAY-HgN4kfEu-A/edit?gid=0#gid=0'
wb = gc.open_by_url(spreadsheet_url)
sheet = wb.get_worksheet(0)

# Load data into DataFrame
data = sheet.get_all_values()
df = pd.DataFrame(data[1:], columns=data[0])

print("Original Data:")
display(df.head())
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd # Import pandas if not already imported in this cell

def phishing_preprocessor(text):
    # Ensure input is string
    text = str(text).lower()

    # 2. Collapse all \n and multiple spaces into one space
    text = re.sub(r'\s+', ' ', text)

    # 3. Replace URLs and Emails
    text = re.sub(r'http\S+', 'URL_TOKEN', text)
    text = re.sub(r'\S*@\S*\s?', 'EMAIL_TOKEN', text)

    # 4. Remove purely numerical junk but keep alphanumeric (like 'requestgene')
    text = " ".join([word for word in text.split() if not word.isdigit()])

    return text.strip()

# Apply the preprocessor to all columns in the dataframe
df_processed = df.copy()
for col in df_processed.columns:
    df_processed[col] = df_processed[col].apply(phishing_preprocessor)

print("Processed Data:")

# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

display(df_processed)
```
